iremos agora separar e adicionar os dados que estão faltando, iremos começar separando as duas populações em jogadores e em Tune squads


```python
#inicia uma variavel para abrigar as info dos players
pop_list = []

for id in player_df['ID']:
    if id <= 30:
        pop_list.append("player"+str(id))
    else:
        pop_list.append('tune_squad'+str(id%30)) #usando id%30, pois senão o id do primeiro tune squad seria 31, usando id%30, irá fazer a divisão e pegar o resto, 31 dividido por 30 é 0 com resto 1
pop_list
```

```python
#atribua esses valores como a coluna "player" no dataframe
player_df['player'] = pop_list
player_df.head()
```
